```python
# importing necessary libraries
import os
import matplotlib.pyplot as plt

import librosa
from librosa.display import waveplot

import IPython.display as ipd

# data path
audio_fpath = ""
audio_clips = ""

audio_test = "audio_file.wav"

# calculating the number of files
# print("Number of audio files: ", os.listdir(audio_clips))

# loading the audio and visualize it
# x, sr = librosa.load(audio_fpath+audio_clips[0], sr=44100) # comment-out this section while using more than one file

x, sr = librosa.load(audio_test, sr=44100)

print(type(x), type(sr))
print(x.shape, sr)#

plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)

# convert audio-form to specrogram
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()

plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()

# save the final image
# plt.savefig('spectrogram.png', dpi=300, bbox_inches='tight') # comment-out when you want to save
```
